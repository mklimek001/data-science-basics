### Financial data analysis with pandas library

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as datareader
```

```python
def fetch_financial_data(company='AMZN'):
    return datareader.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('AAPL')
df.info()  
```

```python
df['avg'] = (df.open + df.close) / 2.0
df.head()
```

```python
df = df.sort_index()
df.head()
```

```python
df['prev_close'] = df.close.shift(1)
df.head()
```

```python
df['daily_change'] = df.close / df.prev_close - 1
df.head()
```

```python
del df['prev_close']
df.head()
```

```python
print("max : ", df.daily_change.max())
print("min : ", df.daily_change.min())
```
